```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
from scipy import stats
```

```python
df = pd.read_csv('miniboone_data.txt',sep=' ',header=None,skiprows=1,skipinitialspace=True)
```

```python
file = open('miniboone_data.txt')
y = file.readline()
file.close()
```

```python
numlabels = [int(i) for i in y.split()]
ylabels = numlabels[0]*[1] + numlabels[1]*[0]
Y = np.array(ylabels)
X = df.to_numpy()
```

```python
dataSignalEventCount = numlabels[0]
dataBackgroundEventCount = numlabels[1]
```

```python
plt.figure()
plt.bar(['Signal Events','Background Events'],[dataSignalEventCount, dataBackgroundEventCount], width = 0.75, color = ['Red', 'Green'])
plt.show()
```

```python
df['Class'] = Y
```

```python
condition = (df.drop(columns = ['Class']) == -999).all(axis=1)
df = df[~condition]
```

```python
correlation = df.corr()
```

```python
df.describe()
```

```python
correlation
```

```python
plt.subplots(figsize = (30,30))
sb.heatmap(correlation, square = True, annot = True)
plt.show()
```

```python
X = df.drop(columns = ['Class']).to_numpy()
Y = df['Class'].to_numpy()
```

```python
plt.figure(figsize = (30, 30))
for i in range(50):
    plt.subplot(10, 5, i+1)
    plt.xlabel('Vrednost')
    plt.ylabel('Ucestalost')
    plt.hist(X[i])
plt.show()
```

```python
plt.figure(figsize = (30, 30))
for i in range(50):
    plt.subplot(10, 5, i+1)
    plt.boxplot(X[i])
plt.show()
```

```python
X = df.drop(columns = ['Class']).to_numpy()
Y = df['Class'].to_numpy()
for i in range(50):
    X[i] = X[i] + abs(X[i].min()) + 1e-8
    X[i],_ = stats.boxcox(X[i])
```

```python
plt.figure(figsize = (30, 30))
for i in range(50):
    plt.subplot(10, 5, i+1)
    plt.xlabel('Vrednost')
    plt.ylabel('Ucestalost')
    plt.hist(X[i])
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize = (30, 30))
for i in range(50):
    plt.subplot(10, 5, i+1)
    plt.xlabel('Vrednost')
    plt.ylabel('Ucestalost')
    plt.hist(X_scaled[i])
plt.show()
```

```python
plt.figure(figsize = (30, 30))
for i in range(50):
    plt.subplot(10, 5, i+1)
    plt.boxplot(X_scaled[i])
plt.show()
```

```python
new_df = pd.DataFrame(X_scaled)
new_df['Class'] = Y
```

```python
new_df.to_csv('miniboone_preprocessed_data.csv', index = False)
```
