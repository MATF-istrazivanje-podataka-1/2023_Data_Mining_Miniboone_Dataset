```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sb
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import AffinityPropagation
from sklearn.datasets import make_blobs
```

```python
df = pd.read_csv("../../analysis_and_preprocessing/miniboone_preprocessed_data.csv")
```

```python
X = df.drop(columns = ['Class'])
Y = df['Class']
```

```python
Y_colors = []
for i in range(Y.shape[0]):
    if Y[i] == 1:
        Y_colors.append('red')
    else:
        Y_colors.append('blue')
```

```python
pca = PCA(n_components = 3)
X_pca = pca.fit_transform(X)
X_pca = pd.DataFrame(X_pca)
```

```python
silhouette_scores = []
sse_scores = []

for k in range(2, 12):
    kmeans = KMeans(n_clusters=k, n_init='auto')
    kmeans.fit(X)
    
    silhouette_score_k = metrics.silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_score_k)
    
    sse_scores.append(kmeans.inertia_)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.plot(range(2, 8), silhouette_scores, marker='o')
ax1.set_xlabel('Number of clusters')
ax1.set_ylabel('Silhouette Score')
ax1.set_title('Silhouette Score for Optimal k')

ax2.plot(range(2, 8), sse_scores, marker='o', color='r')
ax2.set_xlabel('Number of clusters')
ax2.set_ylabel('SSE')
ax2.set_title('Sum of Squared Errors (SSE) for Optimal k')

plt.tight_layout()
plt.show()
```
